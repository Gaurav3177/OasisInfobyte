# Data Preprocessing

## Checking the null values

```python
df.isnull().sum()
```

## Dropping columns CarName and car_ID

```python
df=df.drop(columns=['CarName','car_ID'])
```

```python
df
```

## Using LabelEncoder to convert string values into integer format

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['fueltype'] = le.fit_transform(df['fueltype'])
```

```python
df['aspiration'] = le.fit_transform(df['fueltype'])
df['doornumber'] = le.fit_transform(df['fueltype'])
df['carbody'] = le.fit_transform(df['carbody'])
df['drivewheel'] = le.fit_transform(df['drivewheel'])
df['fuelsystem'] = le.fit_transform(df['fuelsystem'])
df['enginelocation'] = le.fit_transform(df['enginelocation'])
df['enginetype'] = le.fit_transform(df['enginetype'])
df['cylindernumber'] = le.fit_transform(df['cylindernumber'])
```

```python
df
```

### As we can see all the columns are now converted into integer format so we can move furthur

# Splitting the data into train and test

## Removing the columns with very less correlation with price 

```python
x = df.drop(columns=['price','symboling','aspiration','doornumber','carbody','cylindernumber','fueltype'])
y = df[['price']]
```

## We are splitting data in the ratio of 7:3

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

## Checking the cross validation score with different algorithms

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Lasso
from sklearn.svm import SVC
from sklearn.linear_model import Ridge
from sklearn.neighbors import KNeighborsClassifier
```

```python
ridge=cross_val_score(Ridge(), x_test, y_test,cv=9)
np.average(ridge)
```

```python
linear=cross_val_score(LinearRegression(), x_test, y_test,cv=9)
np.average(linear)
```

```python
lasso=cross_val_score(Lasso(), x_test, y_test,cv=9)
np.average(lasso)
```

## As we can see cross val score is maximum in case of Lasso so we will be using this algorithm

# Training the model 

```python
finalmodel=Lasso()
finalmodel.fit(x_test,y_test)
```

```python
finalmodel.score(x_test,y_test)
```
