# Task 1 Iris Classification

```python
import numpy as np
import pandas as pd
```

# Reading the Iris Dataset


## We are using pandas library to read the iris dataset from csv file 

```python
df=pd.read_csv('iris.csv')
df
```

# Data Preprocessing


## Dropping the unnecessary columns

```python
df=df.drop(columns=['Id'])
```

```python
df
```

## Checking for null values

```python
df.isnull().sum()
```

## Using LabelEncoder to convert string values of the column species into integer format

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
```

```python
df
```

# Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns=['Species'])
y = df['Species']
```

```python
x
```

```python
y
```

## We are splitting data in the ratio of 7:3

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_test
```

```python
y_test
```

## Checking the cross validation score with different algorithms 

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Lasso
from sklearn.svm import SVC
from sklearn.linear_model import Ridge
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=cross_val_score(KNeighborsClassifier(n_neighbors=11), x_test, y_test,cv=8)
np.average(knn)
```

```python
logistic=cross_val_score(LogisticRegression(), x_test, y_test,cv=8)
np.average(logistic)
```

```python
ridge=cross_val_score(Ridge(), x_test, y_test,cv=8)
np.average(ridge)
```

```python
svm=cross_val_score(SVC(), x_test, y_test,cv=8)
np.average(svm)
```

```python
linear=cross_val_score(LinearRegression(), x_test, y_test,cv=8)
np.average(linear)
```

```python
lasso=cross_val_score(Lasso(), x_test, y_test,cv=8)
np.average(lasso)
```

## As we can see cross val score is maximum in case of logistic regression so we will be using this algorithm

# Training the model

```python
finalmodel=LogisticRegression()
finalmodel.fit(x_test,y_test)
```

```python
finalmodel.score(x_test,y_test)
```
